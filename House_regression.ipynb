```python
from sklearn.model_selection import train_test_split
```

```python
X = df.drop('price',axis=1).values
```

```python
y = df['price'].values
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
scaled_X_train = scaler.fit_transform(X_train)
```

```python
scaled_X_test = scaler.transform(X_test)
```

```python
#scaled_X_train
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
model = Sequential()

model.add(Dense(512 , activation='relu'))
model.add(Dense(512 , activation='relu'))
model.add(Dense(512 , activation='relu'))
model.add(Dense(20 , activation='relu'))
model.add(Dense(1))

model.compile(optimizer='adam' , loss = 'mse')
```

```python
model.fit(scaled_X_train, y_train, epochs=20, validation_data = (scaled_X_test, y_test))
```

```python
loss = pd.DataFrame(model.history.history)
```

```python
loss
```

```python
loss.plot()
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error
```

```python
y_pred = model.predict(scaled_X_test )
```

```python
mean_absolute_error(y_test,y_pred)
```

```python
new_X = df_without_outliers.drop(['price' , 'zipcode' , 'date'] , axis =1)
```

```python
new_y = df_without_outliers['price']
```

```python
#new_X
```

```python
new_X_train, new_X_test, new_y_train, new_y_test = train_test_split(new_X, new_y, test_size=0.3, random_state=42)
```

```python
scaled_new_X_train = scaler.fit_transform(new_X_train)
```

```python
scaled_new_X_test = scaler.transform(new_X_test)
```

## Apply NN model

```python
new_model = Sequential()

new_model.add(Dense(20 , activation='relu'))


new_model.add(Dense(1))

new_model.compile(optimizer='adam' , loss = 'mse')
```

```python
from tensorflow.keras.callbacks import EarlyStopping
```

```python
early_stop = EarlyStopping(monitor='val_loss' , mode='min' , verbose =1, patience = 50)
```

```python
new_model.fit(scaled_new_X_train,new_y_train,epochs=800 , validation_data=(scaled_new_X_test,new_y_test) , callbacks=[early_stop])
```

```python
new_loss = pd.DataFrame(new_model.history.history)
```

```python
new_loss.plot()
```

```python
new_y_pred = new_model.predict(scaled_new_X_test)
```

```python
mean_absolute_error(new_y_pred , new_y_test)
```
