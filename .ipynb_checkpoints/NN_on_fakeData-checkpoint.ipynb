```python
import tensorflow as tf
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv('fake_reg.csv')
```

```python
df.head()
```

```python
df.shape
```

```python
X = df[['feature1' , 'feature2']].values
```

```python
#X
```

```python
y = df['price'].values
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
X_train.shape
```

## Normalizing/scaling of data

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
scaled_X_train = scaler.fit_transform(X_train)
```

```python
scaled_X_test = scaler.transform(X_test)
```

# Tensorflow 2.0
## Creating models

```python
from tensorflow.keras.models import Sequential
```

```python
from tensorflow.keras.layers import Dense
```

```python
model = Sequential()
model.add(Dense(4 , activation='relu'))
model.add(Dense(4 , activation='relu'))
model.add(Dense(4 , activation='relu'))
model.add(Dense(1))

model.compile(optimizer = 'rmsprop' , loss = 'mse')
```

```python
model.fit(scaled_X_train,y_train , epochs=250)
```

```python
model.history.history
```

```python
loss = model.history.history['loss']
```

```python
plt.plot(loss)
plt.xlabel('Epochs')
plt.ylabel('loss')
```

## comparison

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error
```

```python
model.evaluate(scaled_X_train,y_train)
```

```python
model.evaluate(scaled_X_test,y_test)
```

```python
pred_y = model.predict(scaled_X_test)
```

```python
#pred_y
```

```python
mean_absolute_error(y_test , pred_y)
```

```python
mean_squared_error(y_test ,pred_y)**0.5
```

```python
pred_df = pd.DataFrame(y_test , columns=['y test'])
```

```python
test_pred = pd.Series(pred_y.reshape(300,))
```

```python
pred_df = pd.concat([pred_df , test_pred] , axis = 1)
```
